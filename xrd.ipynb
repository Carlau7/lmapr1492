```python
from pymatgen import Structure
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from jupyter_jsmol.pymatgen import quick_view
from pymatgen.io.vasp import Poscar
from pymatgen.io.cif import CifWriter
from pymatgen.ext.matproj import MPRester
from IPython.display import display, Math 
import numpy as np
from pymatgen.analysis.diffraction.xrd import XRDCalculator
```

```python
#téléchargement du matériau

with MPRester("1dFKRUfD65SqW4BO") as m: 
    # Structure for material id
    structure = m.get_structure_by_material_id("mp-862631")
```

```python
xrd = XRDCalculator('CuKa') #calculateur des diffractions
res = xrd.get_pattern(structure) #résultat des diffractions sur la structure
```

## analyse des pics de diffractions

```python
#premier pic

ang = res.x[0] # l'angle du pic
I = res.y[0] # l'intensité du pic
hkl = res.hkls[0] # les indices hkl du pic

print("Caractéristiques du premier pic du diffractogramme : \n")
print("Angle diffraction: " + str(ang) + "°")
print("Intensité du pic: " + str(I))
print("Indices hkl associés à ce pic: " + str(hkl))
```

```python
#deuxième pic

ang = res.x[1] # l'angle du pic
I = res.y[1] # l'intensité du pic
hkl = res.hkls[1] # les indices hkl du pic

print("Caractéristiques du deuxième pic du diffractogramme : \n")
print("Angle diffraction: " + str(ang) + "°")
print("Intensité du pic: " + str(I))
print("Indices hkl associés à ce pic: " + str(hkl))
```

```python
#troisième pic

ang = res.x[2] # l'angle du pic
I = res.y[2] # l'intensité du pic
hkl = res.hkls[2] # les indices hkl du pic

print("Caractéristiques du troisième pic du diffractogramme : \n")
print("Angle diffraction: " + str(ang) + "°")
print("Intensité du pic: " + str(I))
print("Indices hkl associés à ce pic: " + str(hkl))
```

## visualisation

```python
# diffractogramme global
print("Voyons d'abord un graphe de l'intensité des pics de diffraction du cristal pour des angles allant jusquà 90°")
xrd.show_plot(structure,annotate_peaks=False,two_theta_range=[0,90])
```

```python
#zoom sur les 3 premiers pics
print("Intéressons nous maintenant plus précisément aux trois premiers pics du diffractogramme")
xrd.show_plot(structure,two_theta_range=[0,32])
print("L'influence importante de la multiplicité des triplets (hkl) se remarque bien ici.")
print("En effet, comme indiqué ci-dessus dans les caractéristiques des pics, ceux-ci sont parfois la résultante d'une combinaison des intensités réfléchies de plusieurs plans équivalents.")
print("Cette multiplicité permet notamment de comprendre la valeur beaucoup plus grande de l'intensité du troisième pic")
```
